```python
import os
import time
import requests
import pandas as pd
import urllib3

# (Temporary workaround for SSL issues on some macOS setups)
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# -------------------
# SETTINGS
# -------------------
TOKEN = os.getenv("NOAA_TOKEN")
if not TOKEN:
    raise ValueError("NOAA_TOKEN not found. Restart Jupyter from the same terminal where you exported it.")

BASE_URL = "https://www.ncdc.noaa.gov/cdo-web/api/v2/data"

STATION_ID = "GHCND:USW00014732"   # LaGuardia Airport
START_DATE = "2022-01-01"
END_DATE = "2022-12-31"

# Choose common daily variables (datatype IDs)
DATATYPES = ["TMAX", "TMIN", "PRCP"]

headers = {"token": TOKEN}

def fetch_all_cdo_data(datasetid: str, stationid: str, startdate: str, enddate: str, datatypes=None) -> pd.DataFrame:
    """
    Fetch all NOAA CDO /data results using pagination.
    """
    all_rows = []
    limit = 1000
    offset = 1

    while True:
        params = {
            "datasetid": datasetid,
            "stationid": stationid,
            "startdate": startdate,
            "enddate": enddate,
            "units": "metric",
            "limit": limit,
            "offset": offset
        }

        # Add datatype filters if provided
        if datatypes:
            params["datatypeid"] = datatypes

        r = requests.get(
            BASE_URL,
            headers=headers,
            params=params,
            timeout=60,
            verify=False
        )

        # Helpful debug output if NOAA rejects the request
        if r.status_code != 200:
            print("Status:", r.status_code)
            print("Response:", r.text)

        r.raise_for_status()
        payload = r.json()

        results = payload.get("results", [])
        if not results:
            break

        all_rows.extend(results)

        meta = payload.get("metadata", {})
        resultset = meta.get("resultset", {})
        count = resultset.get("count", 0)

        offset += limit
        if len(all_rows) >= count:
            break

        time.sleep(0.25)

    return pd.DataFrame(all_rows)

# -------------------
# 1) FETCH DAILY WEATHER DATA (2022)
# -------------------
raw = fetch_all_cdo_data(
    datasetid="GHCND",
    stationid=STATION_ID,
    startdate=START_DATE,
    enddate=END_DATE,
    datatypes=DATATYPES
)

if raw.empty:
    raise RuntimeError("No data returned. Double-check station ID, dates, and token.")

# -------------------
# 2) CONVERT TO WIDE FORMAT (one row per date)
# -------------------
raw["date"] = pd.to_datetime(raw["date"]).dt.date

wide = (
    raw.pivot_table(index="date", columns="datatype", values="value", aggfunc="first")
       .reset_index()
       .sort_values("date")
)

# -------------------
# 3) EXPORT TO CSV
# -------------------
output_path = "lga_weather_2022_daily.csv"
wide.to_csv(output_path, index=False)

print(f"✅ Saved {len(wide)} daily rows to: {output_path}")
wide.head()
```

### 5.Merge the weather data with the New York CitiBike data 

```python
# Folder containing CitiBike CSV files
citibike_path = Path(".")

# Collect only CitiBike CSV files
citibike_files = list(citibike_path.glob("202201-citibike-tripdata_*.csv"))

# Read and concatenate
citibike_df = pd.concat(
    (pd.read_csv(f, low_memory=False) for f in citibike_files),
    ignore_index=True
)

print("CitiBike rows:", len(citibike_df))
```

```python
citibike_df["started_at"] = pd.to_datetime(citibike_df["started_at"])
citibike_df["date"] = citibike_df["started_at"].dt.date
```

```python
weather_df = pd.read_csv("lga_weather_2022_daily.csv")

weather_df["date"] = pd.to_datetime(weather_df["date"]).dt.date
```

```python
merged_df = citibike_df.merge(
    weather_df,
    on="date",
    how="left"
)

merged_df.to_csv("citibike_weather_2022.csv", index=False)

print("Merged file saved as citibike_weather_2022.csv")
```
